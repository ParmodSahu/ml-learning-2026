```python
import pandas as pd
import numpy as np
# 1. Create Dataframe
data = {
    'name':['Parmod','Rahul','Bajrang'],
    'age':[32,35,30],
    'salary':[35000,50000,20000]
}
df = pd.DataFrame(data)
print("DataFrame:\n",df)    
```

```python
# 2. Access Data
print("Name:",df['name'].values)
print("First row:", df.iloc[0])
print("Parmod's Salary:", df[df['name'] == 'Parmod']['salary'].values[0])
```

```python
# 3. Statistics
print("Mean Salary:", df['salary'].mean())
print("Max Age:", df['age'].max())
```
