```python
import numpy as np
# 1. Vectors
v1 = np.array([1,2,3])
v2 = np.array([4,5,6])
print("Vector 1:", v1)
print("Vector 2:", v2)
```

```python
# 2. Vector addition
v_sum = v1 + v2
print("Sum:", v_sum)
```

```python
# 3. Scalar multiplication
v_scaled = v1 * 2
print("Scaled by 2:", v_scaled)
```

```python
# 4. Dot product (very important!)
dot_product = np.dot(v1, v2)
print("Dot product:", dot_product)
# Explanation: (1*4) + (2*5) + (3*6) = 4 + 10 + 18 = 32
```

```python
# 5. Vector magnitude (length)
magnitude = np.linalg.norm(v1)
print("Magnitude of v1:", magnitude)
```

```python
# 6. Angle between vectors
cos_angle = np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))
angle = np.arccos(cos_angle) * 180 / np.pi
print(f"Angle between v1 and v2: {angle:.2f} degrees")
```
