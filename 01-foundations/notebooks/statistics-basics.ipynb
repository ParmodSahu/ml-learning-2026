```python
import numpy as np
import matplotlib.pyplot as plt
# Data: Test score of 20 Students
scores = [75, 82, 90, 68, 85, 92, 78, 88, 95, 72,
          80, 88, 76, 89, 91, 67, 84, 79, 93, 86]
# 1. Mean(avarge)
mean = np.mean(scores)
print("Mean:", mean)
```

```python
# 2. Median(middle value)
median = np.median(scores)
print(f"Median:{median}")
```

```python
# 3. Standard Deviation (spread)
std = np.std(scores)
print(f"Standard Deviation: {std}")
```

```python
# 4. Distribution
plt.figure(figsize= (10,6))
plt.hist(scores, bins=8, edgecolor="black")
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.title('Distribution of Test Scores')
plt.axvline(mean, color='red', linestyle='--', label=f'Mean: {mean:.1f}')
plt.axvline(median, color='green', linestyle='--', label=f'Median: {median:.1f}')
plt.legend()
plt.show()
```

```python
# 5. Probability - What % scored above 85?
above_85 = len([s for s in scores if s > 85]) / len(scores) * 100
print(f"Percentage above 85: {above_85:.1f}%")
```
