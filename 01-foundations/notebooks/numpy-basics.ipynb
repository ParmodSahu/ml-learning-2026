```python
import numpy as np
# 1. create arrays
a= np.array([1,2,3,4,5])
print("Array:", a)
print("Shape:", a.shape)
```

```python
# 2. Zero and Ones
zeros = np.zeros((3,4))
ones = np.ones((2,3))
print("Zeros:", zeros)
print("Ones:", ones)
```

```python
# 3. Range
range_arr = np.arange(1,20,4) # 0 to 10, step 2
print("Range:", range_arr)
linspace = np.linspace(0,23,5)
print("Linspace :", linspace)
eyed = np.eye(4)
print(eyed)
random_seed = np.random.randn(3,3)
print(random_seed)
```

```python
# 4. Arithmetic
arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])
print("Addition:", arr1 + arr2)
print("Multiplication:", arr1 * arr2)
print("Element-wise sqrt:", np.sqrt(arr1))
```

```python
# 5. 2D Arrays (metrices)
matrix = np.array([[1,2,3],[4,5,6]])
print("Matrix Shape:", matrix.shape)
print("Matrix:\n", matrix)
```

```python
# 6. Matrix Multiplication
A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])
result = np.dot(A,B) #matrix multiplication
print("Matrix Product \n", result)
```
